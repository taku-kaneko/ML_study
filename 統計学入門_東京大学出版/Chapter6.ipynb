# 第6章　確率分布
気になったことを確認する

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

プロシア騎兵連隊において馬に蹴られて死んだ兵士数が$\lambda = 0.61$のポアソン分布が当てはまるのかを確認

```python
num_death = [0, 1, 2, 3, 4, 5]
num_obs = [109, 65, 22, 3, 1, 0]
num_theory = [108.7, 66.3, 20.2, 4.1, 0.6, 0.1]
df = pd.DataFrame({"死亡数": num_death, "観測数": num_obs, "理論値": num_theory})
df["確率"] = df["観測数"] / df["観測数"].sum()

df
```

```python
def poisson(x, lam=0.61):
    return np.exp(-lam) * (lam**x) / (np.math.factorial(x))
```

```python
df["ポアソン分布"] = [poisson(x) for x in df["死亡数"]]
fig, ax = plt.subplots(figsize=[5, 5])
df.plot(x="死亡数", y=["ポアソン分布", "確率"], ax=ax)
```
