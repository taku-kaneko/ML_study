# 第7章　多次元の確率分布
気になったことを確認する

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

p.139の周辺確率分布

```python
x = y = np.linspace(0, 1, 100)
g_x = 3 * x ** 2
h_y = 3 * (y - 1) ** 2
```

```python
fig, ax = plt.subplots(1, 2, figsize=[8, 4])
ax[0].plot(x, g_x, "-", label=r"$g(x)$")
ax[1].plot(y, h_y, "-", label=r"$h(y)$")
ax[0].legend()
ax[1].legend()
ax[0].set_ylim(0, 4)
ax[1].set_ylim(0, 4)
```
