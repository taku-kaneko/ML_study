# 第4章　練習問題

```python
def pr(r): 
    res = 1
    for i in range(1, r):
        res *= 1 - i / 365
    
    return 1 - res
```

```python
import matplotlib.pyplot as plt

rList = [5, 10, 15, 20, 21, 22, 23, 24, 25, 30, 35, 40, 50, 60]
prList = [pr(r) for r in rList]

fig, ax = plt.subplots()
ax.plot(rList, prList)
ax.set(xlabel=r"$r$", ylabel=r"$p_r$")
```
