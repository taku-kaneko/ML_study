# 第7章　練習問題

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 7.2 iii)
$e_1=0.198, \sigma_1=0.357, e_2=0.055, \sigma_2=0.203, \rho=0.18$のとき，$\mathbb{E}(R_p), V(R_p)$を$x$の関数としてグラフにせよ．

```python
e1 = 0.198
sigma1 = 0.357
e2 = 0.055
sigma2 = 0.203
rho = 0.18
```

```python
x = np.linspace(0, 1, 100)
```

```python
E_Rp = x * e1 + (1 - x) * e2
V_Rp = x ** 2 * sigma1 ** 2 + (1 - x) ** 2 * sigma2 ** 2 + 2 * rho * x * (1 - x) * sigma1 * sigma2
```

```python
fig, ax = plt.subplots()
ax.plot(x, E_Rp, label=r"$\mathbb{E}(R_p)$")
ax.plot(x, V_Rp, label=r"$V(R_p)$")
ax.axhline(y=sigma1 ** 2, c="k", ls="dashed")
ax.axhline(y=sigma2 ** 2, c="k", ls="dashed")
ax.legend(frameon=False)
```
