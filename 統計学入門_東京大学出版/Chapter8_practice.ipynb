```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gaussian(x, mu, sigma2):
    return 1 / np.sqrt(2 * np.pi * sigma2) * np.exp(-(x - mu)**2 / (2 * sigma2))
```

```python
p = 0.4

mu_10 = 10 * (2 * p - 1)
sigma2_10 = 4 * 10 * p * (1 - p)

mu_20 = 20 * (2 * p - 1)
sigma2_20 = 4 * 20 * p * (1 - p)
```

```python
x = np.linspace(-20, 10, 1000)
y_1 = gaussian(x, mu_10, sigma2_10)
y_2 = gaussian(x, mu_20, sigma2_20)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y_1)
ax.plot(x, y_2)
```
