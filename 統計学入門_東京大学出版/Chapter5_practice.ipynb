## 5.1 (ii)
区間$[0, 6]$上の一様分布でチェビシェフの不等式が成立するか
5.1 (i)から$\mu = 3,\quad \sigma = \sqrt{3}$であることがわかっている．
確率は$$P(|X - 3| \geq \sqrt{3}k) = 1 - \frac{2\sqrt{3}}{6}k$$であり，チェビシェフの不等式は$$P(|X - 3| \geq \sqrt{3}k) \leq \frac{1}{k^2}$$である．$k$の範囲は，一様分布の区間と期待値，標準偏差を考えると$0 \leq k \leq \sqrt{3}$の範囲となる．

ここで，$$1 - \frac{2\sqrt{3}}{6}k \leq \frac{1}{k^2}$$を示せれば，チェビシェフの不等式が成立することを示せる．

```python
mu = 3
sigma = np.sqrt(3)
k = np.linspace(0, np.sqrt(3), 100)

prob = 1 - 1 / 3 * np.sqrt(3) * k
Chebyshev = 1 / (k**2)
```

```python
fig, ax = plt.subplots()
ax.plot(k, prob, label="probability")
ax.plot(k, Chebyshev, label="Chebyshev's inequality")
ax.set_ylim(0, 10)
ax.legend()
```

## 5.2 宝くじの期待値

```python
data = np.array([[40000000, 7],
                [10000000, 14],
                [200000, 903],
                [10000000, 5],
                [100000, 645],
                [1000000, 130],
                [140000, 130],
                [10000, 1300],
                [1000, 26000],
                [200, 1300000]])
data = np.vstack([data, [[0, 13000000 - data[:, 1].sum()]]]) # はずれの追加
data = pd.DataFrame(data, columns=["当選金", "本数"])
data
```

```python
data["確率"] = data["本数"] / data["本数"].sum()
```

```python
expected_value = (data["確率"] * data["当選金"]).sum()
print(f"期待値は{expected_value:.2f}")
```

## 5.3 (i)
コインを繰り返し投げ，初めて表が出たときに止める．それが$n$回目であるとき，$2^n$円を得られるものとする．得られる額$X$の確率分布を求めよ．

表が出る確率と裏が出る確率はともに$1/2$とすると，得られる額$X$の確率分布は
$$
f(x) = \sum_i^n \frac{1}{2^n}
$$
となる．ただし，$n = 1, 2, 3, \cdots$

```python
n = np.arange(0, 10 + 1)
prob = (1 / 2)**n
```

```python
fig, ax = plt.subplots()
ax.plot(n, prob)
```
