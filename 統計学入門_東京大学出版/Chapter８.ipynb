```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu = 10
sigma2 = 5
```

```python
n = 100
x = []
for i in range(250):
    r = np.random.rand(n)
    x += [sigma2 * np.sqrt(12 / n) * (r.sum() - n / 2) + mu]
```

```python
fig, ax = plt.subplots()
ax.hist(x)
```
