```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = np.linspace(0.999, 1, 100)
np.log(p / (1 - p))
```

```python
beta0, beta1 = 0, 2
x = np.linspace(-5, 5, 100)
p = 1 / (1 + np.exp(-beta0 - beta1 * x))

fig, ax = plt.subplots()
ax.plot(x, p)
ax.set(title="Logistic dist.")
```

```python
beta = 2
n = 100
mu = 0
sigma2 = 1

x = np.linspace(-10, 10, n)
y = beta * x + np.random.normal(mu, sigma2, n)

beta_list = np.linspace(-5, 5, 100)
loglik = [-n / 2 * np.log(2 * np.pi * sigma2) - 1 / (2 * sigma2) * (y - beta * x) @ (y - beta * x) for beta in beta_list]

fisher_info = -x @ x / sigma2
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

ax = axes[0]
ax.scatter(x, y)

ax.axhline(y=0)
ax.axvline(x=0)

ax = axes[1]
ax.plot(beta_list, loglik)
ax.axvline(x=beta)
ax.axhline(y=0)
ax.set(title=rf"Fisher information: {fisher_info:.2f}")
```
